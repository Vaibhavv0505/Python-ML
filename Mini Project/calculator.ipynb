```python
Expenses =[] #empty list to store all the expenses
while True:

    title = input("Enter expense title: ")
    amount = float(input("Enter amount: "))
    expense= {"title": title, "amount": amount} #create a dictionary for each expense
    Expenses.append(expense) #add the expense dictionary to the list
    choice = input("Do you want to add another expense? (yes/no): ").lower()
    if choice != 'yes':
        break  
#print all expenses
print("\nAll Expenses:")
total = 0
for expense in Expenses:    
    print(f"{expense['title']}: Rs.{expense['amount']}")
    total += expense['amount']
print(f"\nTotal Expenses: Rs.{total}")
```
